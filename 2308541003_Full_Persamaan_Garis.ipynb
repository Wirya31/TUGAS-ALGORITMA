```python
import numpy as np
import matplotlib.pyplot as plt

print("PROGRAM PERSAMAAN GARIS BERDASARKAN NILAI GRADIEN")
print("=================================================")

repetition = True
while repetition:
    kasus_03 = int(input("Kasus berapakah yang akan dikerjakan? (1, 2, atau 3) "))
    
    if kasus_03 == 1:
        print("Masing-masing garis akan melewati dua titik yang diketahui.")
        a_03 = float(input("Nilai x dari titik pertama dari garis pertama adalah ")) # Masukkan nilai x dari titik pertama dari garis pertama
        b_03 = float(input("Nilai y dari titik pertama dari garis pertama adalah ")) # Masukkan nilai y dari titik pertama dari garis pertama
        c_03 = float(input("Nilai x dari titik kedua dari garis pertama adalah ")) # Masukkan nilai x dari titik kedua dari garis pertama
        d_03 = float(input("Nilai y dari titik kedua dari garis pertama adalah ")) # Masukkan nilai y dari titik kedua dari garis pertama

        e_03 = float(input("Nilai x dari titik pertama dari garis kedua adalah ")) # Masukkan nilai x dari titik pertama dari garis kedua
        f_03 = float(input("Nilai y dari titik pertama dari garis kedua adalah ")) # Masukkan nilai y dari titik pertama dari garis kedua
        g_03 = float(input("Nilai x dari titik kedua dari garis kedua adalah ")) # Masukkan nilai x dari titik kedua dari garis kedua
        h_03 = float(input("Nilai y dari titik kedua dari garis kedua adalah ")) # Masukkan nilai y dari titik kedua dari garis kedua

        m1_03 = (d_03 - b_03) / (c_03 - a_03) # Gradien garis pertama
        m2_03 = (h_03 - f_03) / (g_03 - e_03) # Gradien garis kedua

        c1_03 = -m1_03 * a_03 + b_03 # Titik potong garis pertama terhadap sumbu y garis pertama
        c2_03 = -m2_03 * e_03 + f_03 # titik potong garis kedua terhadap sumbu y garis kedua

        print("Persamaan garis pertama adalah y = {:.3f}x + {:.3f}".format(m1_03, c1_03))
        print("Persamaan garis kedua adalah y = {:.3f}x + {:.3f}".format(m2_03, c2_03))

        x_03 = np.linspace(-10, 10, 400)

        if m1_03 == m2_03:
            if c1_03 == c2_03:
                print("Kedua garis berhimpitan sehingga akan terdapat tak berhingga banyaknya solusi")
                plt.plot(x_03, m1_03 * x_03 + c1_03, color='red', label='Garis Berhimpitan')
            else:
                print("Kedua garis saling sejajar")
                print("Kedua garis tidak mungkin berpotongan, sehingga tidak akan terdapat solusi.")
                plt.plot(x_03, m1_03 * x_03 + c1_03, color='red', label='Garis Pertama')
                plt.plot(x_03, m2_03 * x_03 + c2_03, color='blue', label='Garis Kedua')
        else:
            print("Nilai gradien kedua garis berbeda")
            print("Kedua garis berpotongan pada satu titik")
            tp_x_03 = (c2_03 - c1_03) / (m1_03 - m2_03)
            tp_y_03 = m1_03 * tp_x_03 + c1_03
            print("Titik potong kedua garis adalah ({:.3f}, {:.3f})".format(tp_x_03, tp_y_03))
            plt.plot(x_03, m1_03 * x_03 + c1_03, color='red', label='Garis Pertama')
            plt.plot(x_03, m2_03 * x_03 + c2_03, color='blue', label='Garis Kedua')
            plt.plot(tp_x_03, tp_y_03, 'go', label='Titik Potong ({:.3f}, {:.3f})'.format(tp_x_03, tp_y_03))

        plt.axvline(x=0, color='black')
        plt.axhline(y=0, color='black')
        plt.legend(loc='best')
        plt.grid(True)
        plt.show()

    elif kasus_03 == 2:
        print("Masing-masing garis diketahui gradien dan titik yang dilewati.")
        m1_03 = float(input("Gradien dari garis pertama adalah ")) # Masukkan nilai gradien dari garis pertama 
        a_03 = float(input("Nilai x dari titik yang dilewati oleh garis pertama adalah ")) # Masukkan niali x dari titik yang dilewati oleh garis pertama
        b_03 = float(input("Nilai y dari titik yang dilewati oleh garis pertama adalah ")) # Masukkan niali y dari titik yang dilewati oleh garis pertama
        print("Persamaan garis pertama adalah y = {}(x - {}) + {}".format(m1_03, a_03, b_03))
    
        m2_03 = float(input("Gradien dari garis kedua adalah ")) # Masukkan nilai gradien dari garis kedua
        c_03 = float(input("Nilai x dari titik yang dilewati oleh garis kedua adalah ")) # Masukkan niali x dari titik yang dilewati oleh garis kedua
        d_03 = float(input("Nilai y dari titik yang dilewati oleh garis kedua adalah ")) # Masukkan niali y dari titik yang dilewati oleh garis kedua
        print("Persamaan garis kedua adalah y = {}(x - {}) + {}".format(m2_03, c_03, d_03))
    
        c1_03 = -m1_03 * a_03 + b_03
        c2_03 = -m2_03 * c_03 + d_03

        x_03 = np.linspace(-10, 10, 400)

        if m1_03 == m2_03:
            if c1_03 == c2_03:
                print("Kedua garis berhimpitan sehingga akan terdapat tak berhingga banyaknya solusi")
                plt.plot(x_03, m1_03 * (x_03 - a_03) + b_03, color='red', label='Garis Berhimpitan')
            else:
                print("Kedua garis saling sejajar")
                print("Kedua garis tidak mungkin berpotongan, sehingga tidak akan terdapat solusi.")
                plt.plot(x_03, m1_03 * (x_03 - a_03) + b_03, color='red', label='Garis Pertama')
                plt.plot(x_03, m2_03 * (x_03 - c_03) + d_03, color='blue', label='Garis Kedua')
        else:
            print("Kedua garis berpotongan pada satu titik")
            tp_x_03 = (c2_03 - c1_03) / (m1_03 - m2_03)
            tp_y_03 = m1_03 * (tp_x_03 - a_03) + b_03
            print("Titik potong kedua garis adalah ({:.3f}, {:.3f})".format(tp_x_03, tp_y_03))
            plt.plot(x_03, m1_03 * (x_03 - a_03) + b_03, color='red', label='Garis Pertama')
            plt.plot(x_03, m2_03 * (x_03 - c_03) + d_03, color='blue', label='Garis Kedua')
            plt.plot(tp_x_03, tp_y_03, 'go', label='Titik Potong ({:.3f}, {:.3f})'.format(tp_x_03, tp_y_03))

        plt.axvline(x=0, color='black')
        plt.axhline(y=0, color='black')
        plt.legend(loc='best')
        plt.grid(True)
        plt.show()

    elif kasus_03 == 3:
        print("Kedua garis memenuhi persamaan berbentuk ax + by = c dan dx + ey = f.")
        a_03 = float(input("Koefisien dari x pada garis pertama adalah ")) # Masukkan nilai koefisien x pada garis pertama
        b_03 = float(input("Koefisien dari y pada garis pertama adalah ")) # Masukkan nilai koefisien y pada garis pertama
        c_03 = float(input("Konstanta dari garis pertama adalah ")) # Masukkan nilai konstanta dari garis pertama

        d_03 = float(input("Koefisien dari x pada garis kedua adalah ")) # Masukkan nilai koefisien x pada garis kedua
        e_03 = float(input("Koefisien dari y pada garis kedua adalah ")) # Masukkan nilai koefisien y pada garis kedua
        f_03 = float(input("Konstanta dari garis kedua adalah ")) # Masukkan nilai konstanta dari garis kedua

        # Gradient calculation for both lines
        m1_03 = -a_03 / b_03
        m2_03 = -d_03 / e_03

        # Intercept calculation for both lines
        c1_03 = c_03 / b_03
        c2_03 = f_03 / e_03

        x_03 = np.linspace(-10, 10, 400)

        if m1_03 == m2_03:
            if c1_03 == c2_03:
                print("Kedua garis berhimpitan sehingga akan terdapat tak berhingga banyaknya solusi")
                plt.plot(x_03, m1_03 * x_03 + c1_03, color='red', label='Garis Berhimpitan')
            else:
                print("Kedua garis saling sejajar")
                print("Kedua garis tidak mungkin berpotongan, sehingga tidak akan terdapat solusi.")
                plt.plot(x_03, m1_03 * x_03 + c1_03, color='red', label='Garis Pertama')
                plt.plot(x_03, m2_03 * x_03 + c2_03, color='blue', label='Garis Kedua')
        else:
            print("Kedua garis berpotongan pada satu titik")
            A = np.array([[a_03, b_03], [d_03, e_03]])
            B = np.array([c_03, f_03])
            tp_x_03, tp_y_03 = np.linalg.solve(A, B)
            print("Titik potong kedua garis adalah ({:.3f}, {:.3f})".format(tp_x_03, tp_y_03))
            plt.plot(x_03, m1_03 * x_03 + c1_03, color='red', label='Garis Pertama')
            plt.plot(x_03, m2_03 * x_03 + c2_03, color='blue', label='Garis Kedua')
            plt.plot(tp_x_03, tp_y_03, 'go', label='Titik Potong ({:.3f}, {:.3f})'.format(tp_x_03, tp_y_03))

        plt.axvline(x=0, color='black')
        plt.axhline(y=0, color='black')
        plt.legend(loc='best')
        plt.grid(True)
        plt.show()

    else:
        print("Hanya terdapat 3 kasus. Mohon masukkan 1, 2 atau 3 saja.")
    
    jawaban = input("Apakah anda masih ingin mengulang program? (y / n) ")
    if jawaban.lower() == 'n':
        repetition = False
    elif jawaban.lower() == 'y':
        continue
    else:
        print("Tolong masukkan jawaban antara y atau n.")
```
