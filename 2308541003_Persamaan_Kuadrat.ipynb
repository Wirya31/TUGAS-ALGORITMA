```python
import cmath
import matplotlib.pyplot as plt
import numpy as np

persamaan_03=True
while persamaan_03:
    print('PROGRAM PERSAMAAN KUADRAT')
    print('=========================')

    pilihan_03=int(input('Silahkan memilih 2 pilihan program yang ingin anda jalankan: '))
    if pilihan_03==1:
        print('')
        print("PROGRAM MENENTUKAN PERSAMAAN KUADRAT JIKA DIKETAHUI AKAR-AKAR PENYELESAIAN PERSAMAAN KUADRAT")
        print('============================================================================================')
        a1_03=float(input('Masukkan nilai akar penyelesaian 1: '))
        a2_03=float(input('Masukkan nilai akar penyelesaian 2: '))
        
        print('')
        print('Kami akan menentukan jumlah dan hasil kali kedua akar penyelesaian')
        print('Hasil penjumlahan kedua akar penyelesaian')
        print(f'{a1_03} + {a2_03}')
        jumlah_akar_03=a1_03+a2_03
        print(f'Hasil penjumlahan: {jumlah_akar_03}')
        print('')
        print('Hasil kali kedua akar penyelesaian')
        print(f'{a1_03} * {a2_03}')
        kali_akar_03=a1_03*a2_03
        print(f'Hasil kali: {kali_akar_03}')
        print('')
        
        print(f'Persamaan kuadrat dengan akar-akar penyelesaian {a1_03} dan {a2_03} yaitu ')
        print(f'x^2 - {jumlah_akar_03}x + {kali_akar_03}=0')
        print(f'Jadi, persamaan kuadrat dengan akar-akar penyelesaian {a1_03} dan {a2_03} adalah x^2 - {jumlah_akar_03}x + {kali_akar_03}=0')
        print('')
        
        while True:
            ulang_03=input('Apakah anda ingin memulai ulang program (y / t): ')
            if ulang_03.casefold()=='y':
                jwb_03=False
                break
            elif ulang_03.casefold()=='t':
                jwb_03=True
                print('Okee, terima kasih sudah mau menjalankan program ini')
                break
            else:
                print('Silahkan masukkan jawaban yang benar')

        if jwb_03:
            break
        print('')
        
    if pilihan_03==2:
        print('')
        print("PROGRAM MENENTUKAN GRAFIK DAN AKAR-AKAR PENYELESAIAN PERSAMAAN KUADRAT")
        print('======================================================================')
        # Masukkan nilai a,b,c pada persamaan kuadrat
        a_03=float(input('Masukkan nilai a: '))
        b_03=float(input('Masukkan nilai b: '))
        c_03=float(input('Masukkan nilai c: '))

        # Menghitung nilai diskriminan
        D_03=float((b_03**2)-(4*a_03*c_03))
        print('nilai D yaitu: {:.1f}'.format(D_03))

        # Membuat permisalan nilai a berdasarkan 3 kondisi 
        if a_03==0:
            print("Ini bukan persamaan kuadrat, silahkan masukkan nilai a lain.")
            A_03=True
            while A_03:
                a_03=float(input('Masukkan nilai a: '))
                if a_03==0:
                    A_03=True
                elif a_03<0:
                    print("Ini merupakan persamaan kuadrat dengan bentuk grafik parabola terbuka ke bawah.")
                    A_03=False
                else:
                    print("Ini merupakan persamaan kuadrat dengan bentuk grafik parabola terbuka ke atas.")
                    A_03=False

        # Membuat permisalan bentuk grafik berdasarkan nilai a dan diskriminan
        else:
            if a_03<0 and D_03>0:
                print("Ini merupakan persamaan kuadrat dengan bentuk grafik parabola terbuka ke bawah serta memotong sumbu x pada 2 titik.")
            elif a_03<0 and D_03==0:
                print("Ini merupakan persamaan kuadrat dengan bentuk grafik parabola terbuka ke bawah serta memotong sumbu x pada 1 titik.")
                print("Persamaan kuadrat memiliki akar penyelesaian kembar.")
            elif a_03<0 and D_03<0:
                print("Ini merupakan persamaan kuadrat dengan bentuk grafik parabola terbuka ke bawah tanpa mengenai sumbu x.")
                print("Persamaan kuadrat memiliki akar-akar penyelesaian bilangan kompleks.")
            elif a_03>0 and D_03>0:
                print("Ini merupakan persamaan kuadrat dengan bentuk grafik parabola terbuka ke atas serta memotong sumbu x pada 2 titik.")
            elif a_03>0 and D_03==0:
                print("Ini merupakan persamaan kuadrat dengan bentuk grafik parabola terbuka ke atas serta memotong sumbu x pada 1 titik.")
                print("Persamaan kuadrat memiliki akar penyelesaian kembar.")
            else:
                print("Ini merupakan persamaan kuadrat dengan bentuk grafik parabola terbuka ke atas tanpa mengenai sumbu x.")
                print("Persamaan kuadrat memiliki akar-akar penyelesaian bilangan kompleks.")

        # Menentukan titik puncak persamaan kuadrat
        print('')
        xp_03=float(-b_03/(2*a_03))
        print('Nilai sumbu simetri persamaan kuadrat adalah {:.2f}'.format(xp_03))
        yp_03=float(D_03/(-4*a_03))
        print('Nilai optimum persamaan kuadrat adalah {:.2f}'.format(yp_03))
        print(str(f'Jadi, titik puncak persamaan kuadrat y = {a_03}x^2 + {b_03}x + {c_03} adalah' + ' {:.2f}, {:.2f}'.format(xp_03,yp_03)))

        #Menentukan posisi titik puncak persamaan kuadrat
        if xp_03 and yp_03 < 0:
            print('Posisi titik puncak ada di kuadran III.')
        elif xp_03 <= 0 and yp_03 >= 0:
            print('Posisi titik puncak ada di kuadran II.')
        elif xp_03 and yp_03 > 0:
            print('Posisi titik puncak ada di kuadran I.')
        elif xp_03 >= 0 and yp_03 <= 0:
            print('Posisi titik puncak ada di kuadran IV.')
        else:
            print('Posisi titik puncak ada di titik pusat koordinat')

        print('')
        print(f'Persamaan kuadrat adalah y = {a_03}x^2 + {b_03}x + {c_03}') # Mencetak persamaan kuadrat berdasarkan input user

        print('')
        print('Selanjutnya, kami akan membuat grafik berdasarkan bentuk persamaan kuadrat tersebut.')
        print("Diketahui: ")
        print(f"a = {a_03:.2f}")
        print(f'b = {b_03:.2f}')
        print(f'c = {c_03:.2f}')
        print('Pertama-tama kami akan menghitung akar-akar penyelesaian persamaan kuadrat')
        print('substitusi ke dalam rumus (-b ± √(b^2-(4*a*c))) / 2*a')
        # Menghitung akar-akar persamaan kuadrat untuk membentuk grafik 
        if D_03 >= 0:
            x1_03 = (-b_03 + np.sqrt(D_03)) / (2*a_03)
            x2_03 = (-b_03 - np.sqrt(D_03)) / (2*a_03)
        else:
            x1_03 = (-b_03 + cmath.sqrt(D_03)) / (2*a_03)
            x2_03 = (-b_03 - cmath.sqrt(D_03)) / (2*a_03)
            print('Akar-akar penyelesaian persamaan kuadrat tersebut adalah {:.2f} dan {:.2f}'.format(x1_03,x2_03))
            print('')
            print('Berikut adalah grafik persamaan kuadrat tersebut')
            # Membuat rentang nilai x dari -10 hingga 10
            x_03 = np.linspace(-10, 10, 400)

            # Menghitung nilai y yang sesuai untuk persamaan kuadrat
            y_03 = a_03 * x_03**2 + b_03 * x_03 + c_03

            # Membuat plot
            plt.plot(x_03, y_03, label=f"y = {a_03}x^2 + {b_03}x + {c_03}")

            # Menambahkan judul dan label
            plt.title(f"Grafik persamaan kuadrat y = {a_03}x^2 + {b_03}x + {c_03}")
            plt.xlabel("x")
            plt.ylabel("y")

            # Menambahkan kotak dan legenda
            plt.grid(True)
            plt.legend()

            # Menampilkan plot
            plt.show()

            # Mencetak akar persamaan kuadrat
            print("Jadi, akar-akar penyelesaian persamaan kuadrat tersebut adalah ")
            print(f"x = {x1_03:.2f}")
            print(f"x = {x2_03:.2f}")
            print('')
        
        while True:
            ulang_03=input('Apakah anda ingin memulai ulang program (y / t): ')
            if ulang_03.casefold()=='y':
                jwb_03=False
                break
            elif ulang_03.casefold()=='t':
                jwb_03=True
                print('Okee, terima kasih sudah mau menjalankan program ini')
                break
            else:
                print('Silahkan masukkan jawaban yang benar')

        if jwb_03:
            break
        print('')
```
