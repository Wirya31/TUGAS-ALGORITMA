```python
kelompok_03= {} 

while True: 
    grup_03= input('Silahkan masukkan nama kelompok: ') 
    if grup_03 in kelompok_03:
        print("Nama kelompok sudah ada, silahkan masukkan nama kelompok lain")
        print("")
        continue
    else:
        kelompok_03[grup_03] = set()
    
    while True:
        nama_anggota_03= input('Silahkan masukkan nama anggota kelompok: ')
        nama_anggota_sudah_ada_03= False
        for existing_group_03, existing_members_03 in kelompok_03.items():
            if nama_anggota_03 in existing_members_03:
                print(f"{nama_anggota_03} sudah ada di kelompok {existing_group_03}, silahkan masukkan nama lain")
                print("")
                nama_anggota_sudah_ada_03 = True
                continue
            
        if not nama_anggota_sudah_ada_03:
            kelompok_03[grup_03].add(nama_anggota_03)
            
        while True:    
            konfirmasi_03= input('Apakah anda ingin memasukkan nama lagi? ("y" untuk ya, "t" untuk tidak) ')
            if konfirmasi_03.casefold() == 'y':
                isQuit_03 = False
                break
            elif konfirmasi_03.casefold() == 't':
                isQuit_03 = True
                break
            else:
                print("Inputan lain mohon masukkan antara 'y' atau 't' ")
        
        if isQuit_03:
            break
    
    while True:        
        pilihan_03= input('Apakah anda ingin memasukkan grup baru? ("y untuk ya, "t" untuk tidak) ')
        if pilihan_03.casefold() == 'y':
            isQuit_03 = False
            break
        elif pilihan_03.casefold() == 't':
            isQuit_03 = True
            break
        else:
            print("Mohon masukkan input antara 'y' atau 't' ")
            
    if isQuit_03:
        break

print("\nKesimpulan:")
for grup_03, anggota_03 in kelompok_03.items():
    print("\nKelompok", grup_03)
    for anggota_kelompok_03 in anggota_03:
        print(anggota_kelompok_03)
```

```python
belanjaan_03= {}

while True:
    item_03= input("Masukkan item belanjaan (ketik 'selesai' untuk mengakhiri): ")
    if item_03.lower() == 'selesai':
        break
    harga_03= float(input(f"Masukkan harga {item_03}: "))
    jumlah_03= int(input(f"Masukkan jumlah {item_03}: "))
    if item_03 in belanjaan_03:
        belanjaan_03[item_03]["harga"] += harga_03
        belanjaan_03[item_03]["jumlah"] += jumlah_03
    else:
        belanjaan_03[item_03] = {"harga": harga_03, "jumlah": jumlah_03}

print("\nDaftar Belanjaan:")
total_harga_03 = 0
for i_03, (item_03, info_03) in enumerate(belanjaan_03.items(), start=1):
    harga_total_03 = info_03["harga"] * info_03["jumlah"]
    print(f"{i_03}. {item_03}: Rp {info_03['harga']} x {info_03['jumlah']} = Rp {harga_total_03}")
    total_harga_03 += harga_total_03

print("\nTotal Harga Belanjaan:", total_harga_03)
```
