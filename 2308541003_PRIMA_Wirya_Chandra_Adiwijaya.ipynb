```python
def bilangan_prima_03(angka_03):
    if angka_03 <= 1: #Jika input angka <= 1 maka bukan bilangan prima
        return False
    for w_03 in range(2, int(angka_03 ** 0.5) + 1): #Jika terdapat angka yang habis dibagi antara angka 2 hingga akar dari input angka maka bukan bilangan prima 
        if angka_03 % w_03 == 0:
            return False
    return True

prima_03=True #Looping jika user memasukkan angka selain angka 3
while prima_03:
    print('DAFTAR BILANGAN PRIMA ANTARA 1000𝑛 dan 1000(𝑛+1)')
    n_03=int(input('Silahkan masukkan nilai 𝑛 (hanya masukkan angka 3): '))
    if n_03==3:
        prima_03=False
        for a_03 in range(1000*n_03, 1000*(n_03+1)): #Bilangan antara 1000𝑛 dan 1000(𝑛+1) dengan looping 
            if bilangan_prima_03(a_03): #Identifikasi bilangan prima antara 1000𝑛 dan 1000(𝑛+1)
                print(a_03, end=' ') #Cetak bilangan prima antara 1000𝑛 dan 1000(𝑛+1)
    
    else:
        print('Ingat!! 2 digit terakhir NIM Wirya adalah 03, silahkan coba lagi, terima kasih^^')
        print(' ')
```
