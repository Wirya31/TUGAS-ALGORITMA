```python
import numpy as np
import matplotlib.pyplot as plt

# Tambahkan nilai x dan nilai y
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 6, 8, 10])

# Menghitung nilai gradien garis melalui 2 titik
gradient = (y[-1] - y[0]) / (x[-1] - x[0])

# Memunculkan lebih banyak nilai x untuk memetakan suatu garis
x_new = np.linspace(x[0], x[-1], 100)
y_new = gradient * (x_new - x[0]) + y[0]

# Membuat garis dan sumbu
fig, ax = plt.subplots()

# Memetakan garis
ax.plot(x_new, y_new, label=f'y = {gradient:.2f}x + {y[0] - gradient * x[0]:.2f}')

# Memetakan titik-titik asli
plt.scatter(x, y, color='blue')

# Memetakan garis
plt.plot(x_new, y_new, color='red')

# Menambahkan label dan judul
ax.set_ylabel('x')
ax.set_xlabel('y')
ax.set_title('Gradien suatu garis')
ax.legend()

# Menampilkan plot
plt.show()

# Mencetak gradien
print(f'Gradien garis tersebut adalah: {gradient:.2f}')
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the x and y values for the first line
x1 = np.array([1,2,3,4,5])
y1 = np.array([2,4,6,8,10])

# Calculate the gradient of the first line
gradient1 = (y1[-1] - y1[0]) / (x1[-1] - x1[0])

# Generate more x values to plot the first line
x1_new = np.linspace(x1[0], x1[-1], 100)
y1_new = gradient1 * (x1_new - x1[0]) + y1[0]

# Define the x and y values for the second line
x2 = np.array([1,2,3,4,5])
y2 = np.array([3,5,7,9,11])

# Calculate the gradient of the second line
gradient2 = (y2[-1] - y2[0]) / (x2[-1] - x2[0])

# Generate more x values to plot the second line
x2_new = np.linspace(x2[0], x2[-1], 100)
y2_new = gradient2 * (x2_new - x2[0]) + y2[0]

# Create a figure and axis
fig, ax = plt.subplots()

# Memetakan garis
ax.plot(x1_new, y1_new, label=f'y = {gradient:.2f}x + {y1[0] - gradient * x1[0]:.2f}')

# Memetakan titik-titik asli
plt.scatter(x1,y1,color="black")

# Memetakan garis
ax.plot(x2_new, y2_new, label=f'y = {gradient:.2f}x + {y2[0] - gradient * x2[0]:.2f}')

# Memetakan titik-titik asli
plt.scatter(x2,y2,color="green")

# Add labels and title
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Gradien dari dua garis')
ax.legend()

# Display the plot
plt.show()

# Print the gradients
print("Nilai gradien garis pertama adalah:", gradient1)
print("Nilai gradien garis kedua adalah:", gradient2)
```
