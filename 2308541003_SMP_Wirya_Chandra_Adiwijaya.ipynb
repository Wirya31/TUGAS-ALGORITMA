```python
import numpy as np_03
def log_negative(x_03):
    return np_03.log(abs(-x_03))

kuadrat_03=True
while kuadrat_03:
    print('PROGRAM MENGHITUNG PERSAMAAN ALJABAR DENGAN RUMUS PERSAMAAN KUADRAT')
    print('===================================================================')

    print('Diketahui x memenuhi: ')
    print('(x**2) + x - 1 = 0')
    print('')
    print('Tentukan nilai n yang memenuhi')
    print('(x**2024)-(x**2025)=(x*(1-x))**n')
    print('')
    print('penyelesaian:')

    print('Diketahui: ')
    print('a = 1')
    print('b = 1')
    print('c = -1')
    print('')
    print('Pertama-tama kami akan menghitung akar-akar penyelesaian persamaan kuadrat')
    print('substitusi ke dalam rumus (-b ± √(b^2-(4*a*c))) / 2*a')
    
    x_03=str(input('Silahkan masukkan nilai penyelesaian yang ingin ditampilkan (positif/negatif): '))
    print('')

    if x_03=='positif':
        #Nilai x positif
        x1_03 = (-1 + (np_03.sqrt(5))) / 2
        print('Nilai x1 adalah: {}'.format(x1_03))

        # Hitung bagian kiri persamaan
        persamaan_kiri_03 = 2024 * np_03.log(x1_03) + np_03.log(1 - x1_03)

        # Hitung bagian kanan persamaan
        persamaan_kanan_03 = x1_03 * (1 - x1_03)

        # Hitung nilai n positif
        n_positif_03 = persamaan_kiri_03 / np_03.log(np_03.abs(persamaan_kanan_03))
        print(f"Nilai n yang memenuhi: {n_positif_03}")
        print('')
            
        while True:
            extra_03=input('Apakah anda ingin memulai ulang program (y / t): ')
            if extra_03.casefold()=='y':
                respon_03=False
                break
            elif extra_03.casefold()=='t':
                respon_03=True
                print('Okee, terima kasih sudah mau menjalankan program ini')
                break
            else:
                print('Silahkan masukkan jawaban yang benar')

        if respon_03:
            break
        print('')

    elif x_03=='negatif':
        # Nilai x negatif
        x2_03 = (-1 - (np_03.sqrt(5))) / 2
        print('Nilai x2 adalah: {}'.format(x2_03))

        # Hitung bagian kiri persamaan 
        persamaan_kiri2_03 = 2024 * log_negative(x2_03) + np_03.log(1 - x2_03)

        # Hitung bagian kanan persamaan
        persamaan_kanan2_03 = x2_03 * (1 - x2_03)

        # Hitung nilai n negatif
        n_negatif_03 = persamaan_kiri2_03 / log_negative(np_03.abs(persamaan_kanan2_03))
        print(f"Nilai n yang memenuhi: {n_negatif_03}")
        print('')
            
        while True:
            extra_03=str(input('Apakah anda ingin memulai ulang program (y / t): '))
            if extra_03.casefold()=='y':
                respon_03=False
                break
            elif extra_03.casefold()=='t':
                respon_03=True
                print('Okee, terima kasih sudah mau menjalankan program ini')
                break
            else:
                print('Silahkan masukkan jawaban yang benar')

        if respon_03:
            break
        print('')
        
    else:
        print('Silahkan memasukkan jawaban antara "positif" atau "negatif"')
        print('')
```
