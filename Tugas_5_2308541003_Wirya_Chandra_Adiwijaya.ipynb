```python
def fibonacci_03(w_03):
    if w_03==0:
        return 0
    elif w_03==1:
        return 1
    else:
        return fibonacci_03(w_03-1) + fibonacci_03(w_03-2)

def jumlah_fibo_03(w_03):
    sum = 0
    for i_03 in range (w_03):
        sum += fibonacci_03(i_03)
    return sum

deret_03=True
while deret_03:
    print('Program Menghitung Jumlah Deret Bilangan Fibonacci')
    print('==================================================')
    n_03=int(input('Silahkan masukkan angka untuk baris bilangan Fibonacci (0 < bilangan fibonacci < 30): '))
    if 0<=n_03<=30: 
        print(f'Deret bilangan Fibonacci ke-{n_03}')
        for a_03 in range(n_03):
            print(fibonacci_03(a_03))
    else:
        if n_03<0:
            print(f'Angka {n_03} terlalu kecil untuk program ini')
        elif n_03>30:
            print(f'Angka {n_03} terlalu besar untuk program ini')
        else:
            print('Silahkan masukkan angka sesuai ketentuan')
    hasil_03=jumlah_fibo_03(n_03)
    print(f'Jumlah {n_03} deret bilangan fibonacci adalah {hasil_03}')
    print('')
    
    while True:
        ulang_03=input('Apakah anda ingin memulai ulang program (y / t): ')
        if ulang_03.casefold()=='y':
            baris_03=False
            break
        elif ulang_03.casefold()=='t':
            baris_03=True
            print('Terima kasih')
            break
        else:
            print('Silahkan masukkan jawaban yang benar')
            
    if baris_03:
        break
    print('')
```

```python
def display(nama_03, nim_03):
    print("Nama: ", nama_03)
    print("NIM: ", nim_03)
    print()

def search(nim_03, daftar_03):
    for mahasiswa_03 in daftar_03:
        if mahasiswa_03[1] == nim_03: 
            return mahasiswa_03
    return None

def input_mahasiswa_03():
    nama_03 = input("Masukkan nama mahasiswa: ")
    nim_03 = input("Masukkan NIM mahasiswa: ")
    return (nama_03, nim_03)

def checker_mahasiswa_03(daftar_03, nama_03, nim_03):
    nama_mahasiswa_sudah_ada_03 = nim_mahasiswa_sudah_ada_03 = False
    for a_03,r_03 in daftar_03:
        if nama_03 == a_03:
            print(f"{nama_03} sudah ada di dalam data ini, silahkan masukkan nama lain")
            nama_mahasiswa_sudah_ada_03 = True
        if nim_03 == r_03:
            print(f"{nim_03} sudah ada di dalam data ini, silahkan masukkan NIM lain")
            nim_mahasiswa_sudah_ada_03 = True
    return nama_mahasiswa_sudah_ada_03, nim_mahasiswa_sudah_ada_03

print('PROGRAM SEARCHING MAHASISWA')
print('===========================')

daftar_03 = []

jumlah_03 = int(input('Masukkan jumlah mahasiswa: '))

for a_03 in range(jumlah_03):
    mahasiswa_03 = input_mahasiswa_03()
    nama_mahasiswa_sudah_ada_03, nim_mahasiswa_sudah_ada_03 = checker_mahasiswa_03(daftar_03, *mahasiswa_03)
    if not (nama_mahasiswa_sudah_ada_03 or nim_mahasiswa_sudah_ada_03):
        daftar_03.append(mahasiswa_03)
    print('')

nim_search_03 = input("Masukkan NIM yang ingin dicari: ")
hasil_03 = search(nim_search_03, daftar_03)
if hasil_03:
    print("Mahasiswa dengan NIM", nim_search_03, "ketemu")
    display(*hasil_03)
else:
    print("Mahasiswa dengan NIM", nim_search_03, "gak ketemu")
```
