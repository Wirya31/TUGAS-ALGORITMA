```python
kata_03=True
while kata_03:
    print("PROGRAM KALKULATOR SEDERHANA")
    print("============================")
    
    print('1. PENJUMLAHAN')
    print('2. PENGURANGAN')
    print('3. PERKALIAN')
    print('4. PEMBAGIAN')
    print('5. KELUAR')
    
    kalimat_03=int(input('Silahkan masukkan angka 1-5: '))
    if kalimat_03==5:
        kata_03=False
        print('Terima kasih')

    elif int(1<=kalimat_03<=5):
        a=float(input('Masukkan bilangan 1: '))
        b=float(input('Masukkan bilangan 2: '))
        if kalimat_03==1:
            hasil_03=a+b
            print('Hasil penjumlahan adalah ', float(hasil_03))
        elif kalimat_03==2:
            hasil_03=a-b
            print('Hasil pengurangan adalah ', float(hasil_03))
        elif kalimat_03==3:
            hasil_03=a*b
            print('Hasil perkalian adalah ', float(hasil_03))
        elif kalimat_03==4:
            if b!=0:
                hasil_03=a/b
                print('Hasil pembagian adalah ', float(hasil_03))
            else:
                print('Silahkan masukkan angka yang valid')
    elif kalimat_03>5:
        print('Batas akhir yang dapat dimasukkan adalah angka 5')
    else:
        print('Silahkan baca kembali petunjuk program')
    print('')
```

```python
n_03=True
while n_03:
    print('PROGRAM KLASIFIKASI GENERASI')
    print('============================')
    
    tahun_03=int(input('Tahun berapa anda lahir? '))
    if 1946<=tahun_03<=1964:
        print('Anda termasuk dalam generasi Baby Boomer')
        print('Mayoritas generasi anda sudah mempunyai cucu')
    if 1965<=tahun_03<=1980:
        print('Anda termasuk dalam generasi X')
        print('Anda adalah generasi pertama yang menikmati komputer')
    if 1981<=tahun_03<=1995:
        print('Anda termasuk dalam generasi Y')
        print('Anda adalah generasi yang menghabiskan sebagian besar waktu untuk menggunakan gadget')
    if 1996<=tahun_03<=2010:
        print('Anda termasuk dalam generasi Z')
        print('Anda adalah generasi yang sangat peduli dengan kesehatan mental')
    if 2010<=tahun_03<=2024:
        print('Anda termasuk dalam generasi Alpha')
        print('Anda adalah generasi yang dikeliling dengan teknologi')
    
    pertanyaan_03=str(input('Apakah anda ingin melanjutkan program? (ya/tidak)'))
    if pertanyaan_03=='tidak':
        n_03=False
        print('Terima kasih')
    elif pertanyaan_03=='ya':
        n_03=True
        print('')
    else:
        print('Silahkan coba lagi')
        print('')
```
