```python
import numpy as wiryaganteng_03

def validasi_plusminus_matriks_03(matrixA_03, matrixB_03):
    return matrixA_03.shape == matrixB_03.shape

word_03=True

while word_03:
    print("PROGRAM OPERASI MATRIKS SEDERHANA")
    print("=================================")
    
    print('1. PENJUMLAHAN')
    print('2. PENGURANGAN')
    print('3. TRANSPOSE')
    print('4. INVERS')
    print('5. PERKALIAN')
    print('6. KELUAR')
    
    mat_03=int(input('Silahkan anda memilih angka 1-6: '))
    if mat_03==6:
        word_03 = False
        print('Terima kasih')

    elif 1 <= mat_03 <= 5:
        barisA_03 = int(input("Silahkan masukkan jumlah baris matriks A: "))
        kolomA_03 = int(input("Silahkan masukkan jumlah kolom matriks A: "))
        matrixA_03 = wiryaganteng_03.zeros((barisA_03, kolomA_03))
        
        for w_03 in range(barisA_03):
            for a_03 in range(kolomA_03):
                matrixA_03[w_03][a_03] = float(input(f"Enter element ({w_03+1}, {a_03+1}): "))
        print(matrixA_03)
        
        barisB_03 = int(input("Silahkan masukkan jumlah baris matriks B: "))
        kolomB_03 = int(input("Silahkan masukkan jumlah kolom matriks B: "))
        matrixB_03 = wiryaganteng_03.zeros((barisB_03, kolomB_03))
        
        for r_03 in range(barisB_03):
            for i_03 in range(kolomB_03):
                matrixB_03[r_03][i_03] = float(input(f"Enter element ({r_03+1}, {i_03+1}): "))
        print(matrixB_03)
            
        if mat_03==1:
            if validasi_plusminus_matriks_03(matrixA_03, matrixB_03):
                akhir1_03=matrixA_03+matrixB_03
                print(f'Hasil penjumlahan adalah\n')
                print(f'{akhir1_03}')
            else:
                print("Ordo kedua matriks harus sama\n")
        elif mat_03==2:
            if validasi_plusminus_matriks_03(matrixA_03, matrixB_03):
                akhir1_03=matrixA_03-matrixB_03
                print(f'Hasil pengurangan adalah\n')
                print(f'{akhir1_03}')
            else:
                print("Ordo kedua matriks harus sama\n")
        
        elif mat_03==3:
            pil_03=str(input("Silahkan masukkan matriks yang ingin di transpose (A / B): "))
            if pil_03=='A':
                A_transpose_03=wiryaganteng_03.transpose(matrixA_03)
                print(f'tranpose matriks A adalah\n')
                print(f'{A_transpose_03}')
            elif pil_03=='B':
                B_transpose_03=wiryaganteng_03.transpose(matrixB_03)
                print(f'tranpose matriks B adalah\n')
                print(f'{B_transpose_03}')
            else:
                print('Silahkan masukkan jawaban sesuai ketentuan.')
                
        elif mat_03==4:
            pil_03=str(input("Silahkan masukkan matriks yang ingin dilakukan operasi invers (A / B): "))
            if pil_03=='A':
                try:
                    A_invers_03=wiryaganteng_03.linalg.inv(matrixA_03)
                    print(f'invers matriks A adalah\n')
                    print(f'{A_invers_03}')
                except wiryaganteng_03.linalg.LinAlgError:
                    print("Matriks A tidak memiiki invers")
            elif pil_03=='B':
                try:
                    B_invers_03=wiryaganteng_03.linalg.inv(matrixB_03)
                    print(f'invers matriks B adalah\n')
                    print(f'{B_invers_03}')
                except wiryaganteng_03.linalg.LinAlgError:
                    print("Matriks B tidak memiliki invers")
        elif mat_03== 5:
            try:
                hasil_03 = wiryaganteng_03.dot(matrixA_03, matrixB_03)
                print("Hasil perkalian adalah")
                print(hasil_03)
            except ValueError:
                print("Matriks tidak dapat dikalikan")
        print('')
                
    elif mat_03>6 and mat_03<1:
        print('Silahkan masukkan angka antara 1-5')
    else:
        print('Silahkan baca kembali petunjuk program, terima kasih')
print('')
```

```python
import numpy as wiryaganteng_03

def pattern_03(n1_03):
    matrix1_03= wiryaganteng_03.zeros((n1_03, n1_03))
    for y_03 in range(n1_03):
        for r_03 in range(n1_03):
            if y_03 == r_03:
                matrix1_03[r_03, y_03] = 2**y_03
            elif y_03+r_03 == n1_03-1:
                matrix1_03[r_03, y_03] = 2**y_03
    return matrix1_03

n1_03 = int(input("Silahkan masukkan ordo matriks: "))
hasil1_03 = pattern_03(n1_03)
print(hasil1_03)
```

```python
import numpy as wiryaganteng_03

def pola_matriks_03(n_03):
    polamatriks_03 = wiryaganteng_03.zeros((n_03, n_03))
    for w_03 in range(n_03):
        for a_03 in range(n_03):
            if w_03 == 0 or w_03 == n_03-1 or a_03 == 0 or a_03 == n_03-1:
                polamatriks_03[w_03, a_03] = 1
            elif w_03 == a_03:
                polamatriks_03[w_03, a_03] = 2
            elif w_03+a_03 == n_03-1:
                polamatriks_03[w_03, a_03] = 3
    return polamatriks_03

n_03 = int(input("Silahkan masukkan ordo matriks: "))
hasil_03 = pola_matriks_03(n_03)
print(hasil_03)
```
