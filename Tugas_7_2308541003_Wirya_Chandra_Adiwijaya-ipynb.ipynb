```python
import numpy as wiryaganteng_03
import matplotlib.pyplot as wiryakeren_03

# Tambahkan nilai x dan nilai y
x_03= wiryaganteng_03.array([1,2,3,4,5])
y_03= wiryaganteng_03.array([2,4,6,8,10])

# Menghitung nilai gradien garis melalui 2 titik
gradient_03= (y_03[-1] - y_03[0]) / (x_03[-1] - x_03[0])

# Memunculkan lebih banyak nilai x untuk memetakan suatu garis
x_new_03 = wiryaganteng_03.linspace(x_03[0], x_03[-1], 100)
y_new_03 = gradient_03 * (x_new_03 - x_03[0]) + y_03[0]

# Membuat garis dan sumbu
fig_03, ax_03 = wiryakeren_03.subplots()

# Memetakan garis
ax_03.plot(x_new_03, y_new_03, label=f'y = {gradient_03:.2f}x + {y_03[0] - gradient_03 * x_03[0]:.2f}')

# Memetakan titik-titik asli
wiryakeren_03.scatter(x_03, y_03, color='blue')

# Memetakan garis
wiryakeren_03.plot(x_new_03, y_new_03, color='red')

# Menambahkan label dan judul
ax_03.set_ylabel('x')
ax_03.set_xlabel('y')
ax_03.set_title('Gradien suatu garis')
ax_03.legend()

# Menampilkan plot (grafik)
wiryakeren_03.show()

# Mencetak nilai gradien
print(f'Gradien garis tersebut adalah: {gradient_03:.2f}')
```

```python
import numpy as wiryaganteng_03
import matplotlib.pyplot as wiryakeren_03

# Tambahkan nilai x dan y untuk garis pertama 
x1_03 = wiryaganteng_03.array([1,2,3,4,5])
y1_03 = wiryaganteng_03.array([2,4,6,8,10])

# Menghitung nilai gradien garis pertama melalui 2 titik 
gradient1_03 = (y1_03[-1] - y1_03[0]) / (x1_03[-1] - x1_03[0])

# Memunculkan lebih banyak nilai x untuk memetakan garis pertama
x1_new_03 = wiryaganteng_03.linspace(x1_03[0], x1_03[-1], 100)
y1_new_03 = gradient1_03 * (x1_new_03 - x1_03[0]) + y1_03[0]

# Tambahkan nilai x dan y untuk garis kedua
x2_03 = wiryaganteng_03.array([1,2,3,4,5])
y2_03 = wiryaganteng_03.array([3,5,7,9,11])

# Menghitung nilai gradien garis kedua melalui 2 titik 
gradient2_03 = (y2_03[-1] - y2_03[0]) / (x2_03[-1] - x2_03[0])

# Memunculkan lebih banyak nilai x untuk memetakan garis kedua
x2_new_03 = wiryaganteng_03.linspace(x2_03[0], x2_03[-1], 100)
y2_new_03 = gradient2_03 * (x2_new_03 - x2_03[0]) + y2_03[0]

# Membuat garis dan sumbu 
fig_03, ax_03 = wiryakeren_03.subplots()

# Memetakan garis
ax_03.plot(x1_new_03, y1_new_03, label=f'y = {gradient1_03:.2f}x + {y1_03[0] - gradient1_03 * x1_03[0]:.2f}')

# Memetakan titik-titik asli
wiryakeren_03.scatter(x1_03,y1_03,color="black")

# Memetakan garis
ax_03.plot(x2_new_03, y2_new_03, label=f'y = {gradient2_03:.2f}x + {y2_03[0] - gradient2_03 * x2_03[0]:.2f}')

# Memetakan titik-titik asli
wiryakeren_03.scatter(x2_03,y2_03,color="green")

# Menambahkan label dan judul
ax_03.set_xlabel('x')
ax_03.set_ylabel('y')
ax_03.set_title('Gradien dari dua garis')
ax_03.legend()

# Menampilkan plot (grafik)
wiryakeren_03.show()

# Mencetak nilai gradien
print("Nilai gradien garis pertama adalah: ", gradient1_03)
print("Nilai gradien garis kedua adalah: ", gradient2_03)
```
